```sql
-- Datenbank erstellen
CREATE DATABASE Druckerverwaltung;
```

```sql
CREATE TABLE Standort (
    StandortID INT AUTO_INCREMENT PRIMARY KEY,
    Name VARCHAR(255) NOT NULL,
    Adresse VARCHAR(255),
    Stadt VARCHAR(100)
);

CREATE TABLE Mitarbeiter (
    MitarbeiterID INT AUTO_INCREMENT PRIMARY KEY,
    Vorname VARCHAR(100) NOT NULL,
    Nachname VARCHAR(100) NOT NULL,
    StandortID INT,
    FOREIGN KEY (StandortID) REFERENCES Standort(StandortID)
);
```

```sql
CREATE TABLE Drucker (
    DruckerID INT PRIMARY KEY,  
    Modell VARCHAR(100) NOT NULL,  
    Hersteller VARCHAR(100) NOT NULL,  
    Typ VARCHAR(50) NOT NULL,  
    Standort VARCHAR(100),  
    Kaufdatum DATE,  
    Preis DECIMAL(10,2),  
    AbteilungID INT,  
    WartungsvertragID INT  
);
```

```sql
DROP DATABASE IF EXISTS DruckerDB;

CREATE DATABASE DruckerDB;
USE DruckerDB;

CREATE TABLE Abteilung (
    AbteilungID INT PRIMARY KEY,
    Name VARCHAR(100) NOT NULL
);

CREATE TABLE Wartungsvertrag (
    WartungsvertragID INT PRIMARY KEY,
    Anbieter VARCHAR(100) NOT NULL,
    Laufzeit INT NOT NULL
);

CREATE TABLE Drucker (
    DruckerID INT PRIMARY KEY,
    Modell VARCHAR(100) NOT NULL,
    Hersteller VARCHAR(100) NOT NULL,
    Typ VARCHAR(50) NOT NULL,
    Standort VARCHAR(100),
    Kaufdatum DATE,
    KaufPreis DECIMAL(10,2),
    AbteilungID INT,
    WartungsvertragID INT,
    FOREIGN KEY (AbteilungID) REFERENCES Abteilung(AbteilungID),
    FOREIGN KEY (WartungsvertragID) REFERENCES Wartungsvertrag(WartungsvertragID)
);
```
